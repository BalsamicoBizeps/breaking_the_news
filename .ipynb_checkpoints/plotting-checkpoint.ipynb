# Plotting sentiment

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define colors for each sentiment
sentiment_colors = {'positive': 'green',
                    'negative': 'red',
                    'neutral': 'gray'}


# Load the sentiment data for the specific model
file_path = "filtered_sentiment_distribution_by_topic_source_ssary.csv"
sentiment_df = pd.read_csv(file_path, index_col=[0, 1])

# Unique topics and sources
unique_topics = sentiment_df.index.get_level_values('topic').unique()
unique_sources = sentiment_df.index.get_level_values('source').unique()

# Define number of rows and columns for the grid based on number of topics
num_topics = len(unique_topics)
num_sources = len(unique_sources)
fig, axes = plt.subplots(num_topics, num_sources, figsize=(12, 12))
fig.suptitle(f"Sentiment Distribution by Topic and Source for Model {model}", fontsize=16)

# Loop through each topic and source combination to plot
for i, topic in enumerate(unique_topics):
    for j, source in enumerate(unique_sources):
        # Get sentiment distribution for this topic-source pair
        try:
            sentiments = sentiment_df.loc[(topic, source)]
            sentiments = sentiments / sentiments.sum()  # Normalize if necessary
        except KeyError:
            # If no data for this topic-source combination, skip plotting
            continue
        
        # Define colors based on sentiment labels
        colors = [sentiment_colors.get(sentiment, 'blue') for sentiment in sentiments.index]

        # Plotting the donut chart
        ax = axes[i, j]
        wedges, texts, autotexts = ax.pie(
            sentiments,
            labels=sentiments.index,
            autopct='%1.1f%%',
            startangle=90,
            colors=colors,  # Apply custom colors
            wedgeprops=dict(width=0.3)
        )
        
        # Set title to display topic and source
        ax.set_title(f"{topic} - {source}", fontsize=10)
        
        # Improve the layout of each subplot
        for text in texts:
            text.set_fontsize(8)
        for autotext in autotexts:
            autotext.set_fontsize(8)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to fit title
plt.show()
```

# sentiment by source and topic - plotting and statistical analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df = pd.read_csv('/Users/reppmazc/Documents/IRONHACK/quests/final_project/breaking_the_news/filtered_sentiment_distribution_by_topic_source_ssary.csv')

# Step 1: Identify sources that are present in at least 80% of the topics
topic_counts = df['topic'].nunique()
min_topic_count = int(topic_counts * 0.8)

# Count the number of unique topics each source appears in
source_topic_counts = df.groupby('source')['topic'].nunique()
eligible_sources = source_topic_counts[source_topic_counts >= min_topic_count].index

# Filter the DataFrame to include only these eligible sources
df = df[df['source'].isin(eligible_sources)]

# Step 1: Aggregate sentiment proportions by source across all topics
df_agg = df.groupby('source')[['negative', 'neutral', 'positive']].mean().reset_index()

# Step 2: Define colors for each sentiment
sentiment_colors = {'positive': 'green', 'negative': 'red', 'neutral': 'gray'}

def plot_sentiment_by_source_aggregated(df_agg):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.bar(df_agg['source'], df_agg['negative'], color=sentiment_colors['negative'], label='Negative')
    ax.bar(df_agg['source'], df_agg['neutral'], 
           bottom=df_agg['negative'], color=sentiment_colors['neutral'], label='Neutral')
    ax.bar(df_agg['source'], df_agg['positive'], 
           bottom=df_agg['negative'] + df_agg['neutral'], color=sentiment_colors['positive'], label='Positive')
    
    ax.set_title("Average Sentiment Proportions by Source Across All Topics")
    ax.set_xlabel("Source")
    ax.set_ylabel("Average Sentiment Proportion")
    ax.legend()
    plt.xticks(rotation=45)
    plt.show()

# Run the function to plot
plot_sentiment_by_source_aggregated(df_agg)
```

```python
# Step 3: Perform one-way ANOVA for each sentiment type across sources
anova_results = {}
for sentiment in ['negative', 'neutral', 'positive']:
    anova_results[sentiment] = stats.f_oneway(
        *[df[df['source'] == source][sentiment] for source in df['source'].unique()]
    )
    print(f"ANOVA result for {sentiment}: F={anova_results[sentiment].statistic}, p={anova_results[sentiment].pvalue}")

# Step 4: Post-hoc Testing with Tukey’s HSD (if ANOVA indicates significant differences)
for sentiment in ['negative', 'neutral', 'positive']:
    if anova_results[sentiment].pvalue < 0.05:  # Check if ANOVA is significant
        tukey = pairwise_tukeyhsd(df[sentiment], df['source'])
        print(f"Tukey's HSD result for {sentiment} sentiment:")
        print(tukey)
```
